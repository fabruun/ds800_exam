Extend `text.py` with the following functions:
- **Word context analysis:** *"The context of a word [...] consists of the words[...] before and after it which help to make its meaning clear.‚Äù* Write a function `context(text, word)` that counts the words occurring in the context of word. For simplicity, the context of a word shall be defined by the words placed on the same line as any occurrence of word. The function should return a dictionary that is used for counting the context words.
- **Synonymous words:** Write a function `synonymous(text, w1, w2)` that compares the contexts of two words `w1` and `w2` in the given text with each other. The function should return a value in range [0,1] that indicates how synonymous the two words are.
- **Word frequency change:** Write a function `change(text, word)` that visualizes the changing frequency of word in the text. Therefore, count for each 100 lines of text how often word appears and append the count to a list. Visualize the changing frequency with a line chart!

```python
def context(text, word):
    context_words = {}
    
    f = open(f"./texts/{text}", "r", encoding="utf-8")
    lines = f.readlines()
    for line in lines:
        if word in line:
            words = line.split()
            for w in words:
                if w != word:
                    if w in context_words.keys():
                        context_words[w] += 1
                    else:
                        context_words[w] = 1
                        
    return context_words
        
print(context("Lord of the rings", "ring"))
```

```python
def synonymous(text, w1, w2):
    
    s1 = []
    s2 = []
    synonymous_words = []
    f = open(f"./texts/{text}", "r", encoding="utf-8")
    lines = f.readlines()
    
    for line in lines:
        if w1 in line:
            words = line.split()
            for word in words:
                if w1 != word:
                    s1.append(word)
        elif w2 in line:
            words = line.split()
            for word in words:
                if w2 != word:
                    s2.append(word)
    
    for item in s1:
        for val in s2:
            if item == val:
                synonymous_words.append(item)
    
    if len(synonymous_words) >= len(s1+s2):
        return len(s1+s2)/len(synonymous_words)
    else:
        return len(synonymous_words)/len(s1+s2)
    
synonymous("Frankenstein", "edamame", "sushi")
```

```python
import matplotlib.pyplot as plt

def change(text, word):
    frequency = []
    count = 0
    f = open(f"./texts/{text}", "r", encoding="utf-8")
    lines = f.readlines()
    for line in lines:
        if lines.index(line) % 100 == 0:
            frequency.append(count)
            count = 0
        else:
            if word in line:
                count += 1
    
    plt.plot(frequency)

change("Frankenstein", "God")
```
