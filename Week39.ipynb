# Exercise 1
Write  a  function `aristocrats(persons)` that  returns  a  list  of  persons  who have an " af "," von "," van " or " de " in their name!

```python
# Import json library for handling json
import json

def aristocats(persons):
    # define names of aristocrats in a list
    names = ["af", "von", "van", "de"]
    # define empty list for aristocrat in the dbl
    aristocats = []
    
    # Open the json file and load the data
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
    
    # loop through the persons in the dbl
    for d in data:
        # loop through the names in the list
        for name in names:
            # check if any of the names appear in dbl
            if name in d["name"]:
                # if they appear add them to the aristocrats list
                aristocats.append(d["name"])
    
    # return the aristocrats list
    return aristocats

print(aristocats("dbl.json"))
```

# Exercise 2
Write  a  function `first_name(persons, name)` that  returns  a  list  of  persons who have a specific first name name.  Test your function by printing all persons who have the same first name as you.

```python
# Import json for handling json
import json

def first_name(persons, name):
    # Define a empty list for people with same first name as name
    people = []
    
    # Open the json file and file load the data
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
    
    # loop the data
    for d in data:
        # check if name is in the data
        if name + " " in d["name"]:
            # add person to the list
            people.append(d["name"])
            
    return people

print(first_name("dbl.json", "Hans"))
```

# Exercise 3
Write a function `born_rural(persons)` that returns a list of persons who have not been born in either Copenhagen, Aarhus or Odense!

```python
import json

def born_rural(persons):
    cities = ["Copenhagen", "Odense", "Aarhus"]
    rural_people = []
    
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
        
    for d in data:
        if d["p"] not in cities:
            rural_people.append(d["name"])
            
    return rural_people

print(born_rural("dbl.json"))
```

# Exercise 4
Write a function `older_than(persons, age)` that returns a list of persons who reached an age older than age.  Test your function by printing all persons who reached an age of at least 90 years!

```python
import json
import datetime

def older_than(persons, age):
    old_people = []
    
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
        
    year = datetime.datetime.now().year
    
    for d in data:
        if d["l"][0] != "-" and year - int(d["l"][0:4]) >= 90:
            old_people.append(d["name"])
            
    return old_people
    
print(older_than("dbl.json", 90))
```

# Exercise 5
Write a function `death_between(persons, year_min, year_max)` that returns a list of persons who died between year_min and year_max. Test your function by printing all persons who died between 1940 and 1945!

```python
import json

def death_between(persons, year_min, year_max):
    dead_people = []
    years = []
    for i in range(year_min, year_max + 1):
        years.append(i)
    
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
        
    for d in data:
        year_died = d["l"].split("-")
        if len(year_died[1]) > 0:
            if int(year_died[1]) in years:
                dead_people.append(d["name"])
                
    return dead_people
        
print(death_between("dbl.json", 1940, 1945))
```

# Exercise 6
Write a function `painters(persons)` that returns a list of all painters who had their mean year of activity in the Danish golden age between 1800 and 1850!

```python
import json

def painters(persons):
    years = []
    people = []
    
    for i in range(1800, 1851):
        years.append(i)
        
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
        
    for d in data:
        if "painter" in d["o"] and d["y"] in years:
            people.append(d["name"])
            
    return people
            
print(painters("dbl.json"))
```

# Exercise 7 
Write a function `occupations(persons, olist)` that returns a list of persons that had all occupations listed in olist.  Test your function by printing all persons who followed the occupations ”writer” and ”judge”!

```python
import json

def occupations(persons, olist):
    people = []
    
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
        
    for d in data:
        for o in d["o"]:
            if o in olist:
                people.append(d["name"])
                
    return people

print(occupations("dbl.json", ["writer", "judge"]))
```

# Exercise 9
Write  a  function `incomplete(persons)` that returns a list containing three lists:  the first one contains all persons without years of birth and death, the second contains all persons with only year of birth, and the third contains all persons with only year of death!

```python
import json

def incomplete(persons):
    people = []
    no_year = []
    no_birth = []
    no_death = []
    
    with open(persons, encoding="utf-8") as json_file:
        data = json.load(json_file)
        
    for d in data:
        if d["l"] == "-":
            no_year.append(d["name"])
        elif d["l"][-1] == "-":
            no_birth.append(d["name"])
        elif d["l"][0] == "-":
            no_death.append(d["name"])
            
    people.append(no_year)
    people.append(no_birth)
    people.append(no_death)
    
    return people

print(incomplete("dbl.json"))
```
